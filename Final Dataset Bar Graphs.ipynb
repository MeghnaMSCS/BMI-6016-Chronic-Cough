```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the csv file. Takes a long time even with CHPC
cc_final = pd.read_csv("chronic_cough_final.csv",engine='python')
```

```python
#Convert to pickle file for slight compression
cc_final.to_pickle("chronic_cough_final.pkl")
```

```python
#Read pickle file
cc_final = pd.read_pickle('chronic_cough_final.pkl')
```

```python
cc_final
```

```python
#Get number of chronic cough encounters grouped by race
cc_final2 = cc_final.groupby(['race'])['c_cough_yn'].sum()
```

```python
cc_final2
```

```python
#Create bar graph to visualize data. 
ax = cc_final2.plot.bar()
ax.set_xlabel('Race')
ax.set_title('Chronic Cough Instances per Race')
ax.ticklabel_format(style='plain',axis='y')
ax.bar_label(ax.containers[0]);
```

```python
#Get number of chronic cough encounters grouped by sex
cc_final3 = cc_final.groupby(['sex'])['c_cough_yn'].sum()
```
